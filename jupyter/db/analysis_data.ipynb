```python
def get_len(tb):
    '''
    테이블 칼럼별 최대 길이 조회
    '''
    max_lens = dict()
    cols = tb.columns
    for col in cols:
        if tb[col].dtype == 'object':
            count_df = tb[col].copy()
            count_df = pd.DataFrame(count_df, columns=[col])
            count_df.dropna(inplace=True)
            max_lens[col] = max(count_df[col].str.len())
    return max_lens 
```
